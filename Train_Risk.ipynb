```python
u, count = np.unique(HF['pre_internal_PatientAnonId'], return_counts=True)
count_sort_ind = np.argsort(-count)
d = dict(zip(u[count_sort_ind],count[count_sort_ind]))

for id in d.keys():
    x = HF[HF['pre_internal_PatientAnonId']==id]['pre_internal_VisitAnonId']
    if len(x)>1:
        print('Dropping duplicates for id', id)
        #keep the last visit ID
        HF.drop(index=x.index[:-1],inplace=True)
        
#verify no duplicate entries
np.unique(HF['pre_internal_PatientAnonId'], return_counts=True)
```
